```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear el DataFrame con los nuevos datos
data = [
    (1, 'Dinamarca', 90),
    (2, 'Finlandia', 88),
    (3, 'Singapur', 84),
    (4, 'Nueva Zelanda', 83),
    (5, 'Luxemburgo', 81),
    (5, 'Noruega', 81),
    (5, 'Suiza', 81),
    (8, 'Suecia', 80),
    (9, 'Países Bajos', 78),
    (10, 'Australia', 77),
    (10, 'Islandia', 77),
    (10, 'Irlanda', 77),
    (13, 'Estonia', 77),
    (13, 'Uruguay', 77),
    (15, 'Canadá', 75),
    (15, 'Alemania', 75),
    (17, 'Hong Kong', 74),
    (18, 'Bután', 72),
    (18, 'Seychelles', 72),
    (20, 'Japón', 71),
    (20, 'Reino Unido', 71)
]

# Crear el DataFrame
df_transparencia = pd.DataFrame(data, columns=['Rango', 'País', 'Índice de Transparencia'])

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
bars = plt.bar(df_transparencia['País'], df_transparencia['Índice de Transparencia'], color='lightblue')

# Resaltar la barra de Uruguay
for bar in bars:
    if bar.get_x() == bars[df_transparencia[df_transparencia['País'] == 'Uruguay'].index[0]].get_x():
        bar.set_color('orange')  # Cambiar el color de la barra de Uruguay

# Añadir etiquetas y título
plt.xlabel('Países')
plt.ylabel('Índice de Transparencia')
plt.title('Índice de Transparencia de los Países')
plt.xticks(rotation=45)
plt.ylim(0, 100)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the countries
countries = ['Ecuador', 'Uruguay']
scores = [32, 77]
rankings = [121, 13]

# Create a figure and axis
fig, ax1 = plt.subplots(figsize=(8, 5))

# Create a bar chart for scores
bars = ax1.bar(countries, scores, color=['lightblue', 'orange'], alpha=0.7)

# Add labels and title
ax1.set_xlabel('Países')
ax1.set_ylabel('Índice de Transparencia', color='black')
ax1.set_title('Comparativa del Índice de Transparencia: Ecuador vs Uruguay')
ax1.set_ylim(0, 100)

# Annotate the scores on the bars
for bar in bars:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval + 1, yval, ha='center', va='bottom')

# Create a second y-axis for rankings
ax2 = ax1.twinx()
ax2.set_ylabel('Ranking', color='black')
ax2.set_ylim(0, 130)

# Create a line plot for rankings
ax2.plot(countries, rankings, color='red', marker='o', label='Ranking', linewidth=2)

# Annotate the rankings on the line
for i, rank in enumerate(rankings):
    ax2.text(i, rank + 1, rank, ha='center', va='bottom', color='red')

# Show the legend
ax2.legend(loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```
